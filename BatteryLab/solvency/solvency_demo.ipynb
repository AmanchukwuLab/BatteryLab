Objective: given a bank of solvents with compositions $$\vec{x}_e = [v_i, s_i, a_i]$$ where $v_i$ are volume fractions, $s_i$ are salt concentrations (mol/L), and $a_i$ are additive concentrations (mol/L), as well as a desired final solvent composition to be made by mixing available solvents, create a solver to identify the individual volume amounts $V_e$ for each electrolyte from the bank to compose the final electolyte.

```python
from pathlib import Path
import sys
import numpy as np

# Add the repo root to sys.path so imports work when
# the notebook is opened directly from this project folder.
repo_root = Path.cwd().resolve().parent
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

from solvency.core import Electrolyte, e_solver

local_smiles = {
    "water": "O",
    "propylene glycol": "CC(O)CO",
    "hexane": "CCCCCC",
    "NaCl": "[Na+].[Cl-]",
    "H2SO4": "OS(=O)(=O)O",
}
```

# Package Demo

```python
# Build a tiny in-notebook demo using the packaged API.
e1 = Electrolyte(
    name="saltwater-propylene glycol-acid mix",
    volume=150,
    v={"water": 0.5, "propylene glycol": 0.5},
    s={"NaCl": 2.0},
    a={"H2SO4": 0.1},
    local_smiles=local_smiles,
)

e2 = Electrolyte(
    name="water",
    volume=500,
    v={"water": 1.0},
    local_smiles=local_smiles,
)

e3 = Electrolyte(
    name="saltwater",
    volume=200,
    v={"water": 1.0},
    s={"NaCl": 5.0},
    local_smiles=local_smiles,
)

e4 = Electrolyte(
    name="hexane",
    volume=500,
    v={"hexane": 1.0},
    local_smiles=local_smiles,
)

e5 = Electrolyte(
    name="2M sulfuric acid", 
    volume=100,
    v={"water": 1.0},
    s={"H2SO4": 2.0},
    local_smiles=local_smiles,
)

bank = [e1, e2, e3, e4, e5]
final = Electrolyte(
    name="goal electrolyte",
    volume=50,
    v={"water": 10, "hexane": 0.01},
    s={"NaCl": 3.0},
    a={"H2SO4": 0.05},
    local_smiles=local_smiles,
)
```

```python
solution = e_solver(bank, final)
print("solution:", solution)

# Sum all values in 'solution'
for name, vol in solution.items():
    print(f"{name}: {vol*50}")
```
